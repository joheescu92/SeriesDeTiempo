```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from google.colab import files
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/data/tienda_ventas.csv')
df
```

```python
#agrapación
por_dias= df[['date','sales']].groupby('date').sum()
por_dias
#gráfico de líneas
por_dias.plot.line(
    y ='sales',
    figsize=(30,10)
)
plt.title('Ventas por Días')
```

```python
#Gráfico en Plotly
fig= px.line(
    por_dias,
    x = por_dias.index.values,
    y = 'sales',
    title='Ventas por Días'
)
fig.show()
```
